### Question 6: Label Mapping (5 Marks)

Since K-Means does not know the actual digit labels, we need to map our cluster labels (0-9) to the most common true digit label within each cluster. Import `scipy.stats.mode`. For each cluster, find the mode of the true `digits.target` labels. Store this mapping in a dictionary called `labels_map`.

**Hint:** Loop from cluster 0 to 9. In each loop, find the mode of `digits.target[clusters == i]` where `i` is the cluster number.

**Expected Output:** A dictionary where keys are cluster labels and values are the corresponding most frequent true digit labels.

````python
# Import required libraries
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from scipy.stats import mode

# Load the dataset
digits = load_digits()

# Create PCA model and reduce to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(digits.data)

# Perform K-Means clustering on reduced_data
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(reduced_data)

# Create dictionary to map cluster labels to true digit labels
labels_map = {}

# Loop from cluster 0 to 9
for i in range(10):
    # Find the mode of true digits.target labels for cluster i
    cluster_targets = digits.target[clusters == i]
    mode_result = mode(cluster_targets)
    # Store the most frequent true digit label (use .item() to safely extract the scalar value)
    labels_map[i] = mode_result.mode.item()

# Print the labels_map dictionary
print(labels_map)
````

### Question 7: Creating Final Predictions (3 Marks)

Using the `labels_map` dictionary from the previous step, create an array of predicted labels `pred_labels` by converting your K-Means `clusters` array to the mapped digit labels. Print the first 10 predicted labels.

**Expected Output:** An array showing the first 10 final predicted digit labels.

````python
# Import required libraries
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from scipy.stats import mode
import numpy as np

# Load the dataset
digits = load_digits()

# Create PCA model and reduce to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(digits.data)

# Perform K-Means clustering on reduced_data
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(reduced_data)

# Create the labels_map dictionary (from previous step)
labels_map = {}
for i in range(10):
    cluster_targets = digits.target[clusters == i]
    mode_result = mode(cluster_targets)
    labels_map[i] = mode_result.mode.item() # Fix: use .item() to extract scalar value

# Create pred_labels by converting clusters array to mapped digit labels
pred_labels = np.array([labels_map[cluster] for cluster in clusters])

# Print the first 10 predicted labels
print(pred_labels[:10])
````

### Question 8: Evaluate Clustering Accuracy (2 Marks)

Import `accuracy_score` from `sklearn.metrics`. Calculate and print the accuracy of your unsupervised clustering by comparing the `pred_labels` with the true `digits.target` labels.

**Expected Output:** A single floating-point number representing the accuracy score.

````python
# Import required libraries
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from scipy.stats import mode
import numpy as np
from sklearn.metrics import accuracy_score

# Load the dataset
digits = load_digits()

# Create PCA model and reduce to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(digits.data)

# Perform K-Means clustering on reduced_data
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(reduced_data)

# Create the labels_map dictionary
labels_map = {}
for i in range(10):
    cluster_targets = digits.target[clusters == i]
    mode_result = mode(cluster_targets)
    labels_map[i] = mode_result.mode.item() # Fix: use .item() to extract scalar value

# Create pred_labels by converting clusters array to mapped digit labels
pred_labels = np.array([labels_map[cluster] for cluster in clusters])

# Calculate and print the accuracy
accuracy = accuracy_score(digits.target, pred_labels)
print(accuracy)
````

### Question 9: Confusion Matrix (2 Marks)

Import `confusion_matrix` from `sklearn.metrics`. Compute and print the confusion matrix to see which digits were most often confused with each other.

````python
# Import required libraries
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from scipy.stats import mode
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# Load the dataset
digits = load_digits()

# Create PCA model and reduce to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(digits.data)

# Perform K-Means clustering on reduced_data
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(reduced_data)

# Create the labels_map dictionary
labels_map = {}
for i in range(10):
    cluster_targets = digits.target[clusters == i]
    mode_result = mode(cluster_targets)
    labels_map[i] = mode_result.mode.item() # Fix: use .item() to extract scalar value

# Create pred_labels by converting clusters array to mapped digit labels
pred_labels = np.array([labels_map[cluster] for cluster in clusters])

# Compute and print the confusion matrix
cm = confusion_matrix(digits.target, pred_labels)
print(cm)
````

### Question 10: Model Evaluation with Classification Report (2 Marks)
